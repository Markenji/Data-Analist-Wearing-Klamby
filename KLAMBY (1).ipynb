## Data Cleaning

### Checking Missing Value

```python
df.isna().sum()
```

Tidak terdapat missing value pada dataset diatas.

### Checking Duplicates

```python
# Check for Duplicates
df[df.duplicated()]
```

Tidak terdapat duplicates pada dataset diatas.

### Change data type

```python
#Merubah tipe data pada kolum yang salah
df['gross_sales'] = [int(str(i).replace(".", "")) for i in df['gross_sales']]
df['discounts'] = [int(str(i).replace(".", "")) for i in df['discounts']]
df['returns'] = [int(str(i).replace(".", "")) for i in df['returns']]
df['net_sales'] = [int(str(i).replace(".", "")) for i in df['net_sales']]
df['total_sales'] = [int(str(i).replace(".", "")) for i in df['total_sales']]
```

### Memastikan Perubahan Data

```python
df.info()
```

Pada dataset diatas dapat dilihat bahwa tipe data pada kolum ('gross_sales', 'discounts', 'returns', 'net_sales',  'total_sales', 'net_quantity') sudah memiliki tipe data yang sesuai

```python
df = pd.get_dummies(df, columns=['referrer_source'])
```

```python
df
```

```python
df.columns
```

```python
num_features = df[['orders', 'gross_sales', 'discounts', 'returns', 'net_sales', 'shipping', 'total_sales', 'net_quantity']]
```

```python
# Visualisasi
count=0
fig = plt.figure(figsize=(15, 20))
palette = ['plum', '#aa0022']
sns.set(palette='bright')
categorical_feature= num_features
for feature in categorical_feature:
    count +=1
    ax= plt.subplot(6,2,count)
    sns.histplot(data=df,x=feature,palette=palette)
    plt.title("Plot for {}".format(feature))
    fig.tight_layout()
plt.show()
```

Pada grafik histogram didapatkan bahwa grafik yang terdistribusi normal adalah net_sales dan total_sales.

```python
num_features = df[['orders', 'gross_sales', 'discounts', 'returns', 'net_sales', 'shipping', 'total_sales', 'net_quantity']]
```

```python
# Visualisasi
count=0
fig = plt.figure(figsize=(15, 20))
palette = ['plum', '#aa0022']
sns.set(palette='bright')
categorical_feature= num_features
for feature in categorical_feature:
    count +=1
    ax= plt.subplot(6,2,count)
    sns.boxplot(data=df,y=feature,palette=palette)
    plt.title("Plot for {}".format(feature))
    fig.tight_layout()
plt.show()
```

Pada gambar diatas terlihat bahwa terpadat 3 kolum dengan otulier yang tinggi pada total_sales, net_sales, dan gross_sales

## Kesimpulan dan Rekomendasi

<b>Dari analisis yang telah dilakukan, kita bisa membuat kesimpulan tentang faktor faktor yang bisa di optimalkan oleh Klamby:</b>
- Referrer_source paling banyak berasal dari iklan pada media sosical yang di klik oleh customer sehingga perlu memperhatikan dan melakukan optimisasi iklan dengan sebaik mungkin.
- Karena Referrer_source adalah faktor yang penting untuk mendapatkan customer maka perlu mengikuti perkembanga jaman yaitu melakukan iklan pada platfrom terkenal saat ini seperti Tiktok, Adsen, dan Instagram.
- Pada produk yang didiscount contoh saja Valerie Dress Powder Blue sebanyak 100 juta lebih baik produksinya dikurangi dan discount disebar rata kesemua produk.
- Selena Dress Hazel adalah produk dengan pendapatan gross sales dan total_sales terbesar sekitar 2,85 Miliar sehingga produk ini coba dibuat versi yang di update kedepanya sehingga mencari peluang laku lebih tinggi. Serta diberikan discount pada produk ini.
- Kota dengan shipping terbanyak adalah Tangerang Selatan sebanayak 9 sehingga perlu dijaga hubungan customer disana perlu untuk memberikan discount untuk pembelian kedepanya.
- Setelah penjualan dalam negeri mencapai target minimum makan akan lebih baik untuk mencoba pasar seperti negara tetangga seperti malaysia.
